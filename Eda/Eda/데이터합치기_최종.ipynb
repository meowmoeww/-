```python
#녹조정보를 NaN값 제외 원하는 데이터프레임 형식으로 만들기

#1. 대교 이름 - 날짜만큼 복사
def cc(df):
  #공통으로 사용할 변수
  br=df.columns[1:]

  bridge=[] #대교이름만 빼기
  for i in br:
    for j in range(df.shape[0]-df.isna().sum()[i]): #그 대교의 일시isna 수만큼 ->  (나중에 일자isna빼고 세줘야 함.), 아 이것도 그 대교 수만큼 해줘야 하네.......
      bridge.append(i)
  bd=pd.DataFrame(bridge)

  #2.일자
  #위에처럼 가면 안될듯 열별로해서 그 행 isna인지 확인하고 그 일자 넣어줘야 함.
  #대교별 녹조를 확인해서 isna가 아니면 그 때의 loc추가
  #index떼면 쉬울isna인 애들 index 뗀다 -> loc
  #결과값 date

  date=[]
  br=df.columns[1:] #미리 함수 밖에 빼놓는 게 나을듯

  for i in br:
    nan_dex=df[~df[i].isna()].index
    date.extend(df['조사일자'].loc[nan_dex])
  date=pd.DataFrame(date)

  #녹조(nan이 아니면 밑으로 쭉 붙이기)

  nok=[]
  for i in br:
    nan_dex=df[~df[i].isna()].index
    nok.extend(df[i].loc[nan_dex])
  nokjo=pd.DataFrame(nok)

  df=pd.concat([bd,date,nokjo],axis=1)
  df.columns=['대교','일자','녹조']

  #pd.cut으로 경보, 위험도 같이 넣기
  #상수원,친수구역
  u=['미사대교','강동대교','잠실교','광진교']
  d=['성수대교','한강대교','한남대교','마포대교','성산대교']
  df.loc[df['대교'].isin(u),'경보']=pd.cut(df['녹조'],bins=[-1,1000,10000,1000000,1000000000],labels=['정상','관심','경계','대발생']) #첫 단위 이후부터 세서 -1로 산정, 맨 마지막 임의로 넘어줌.


  return df
```

```python
#원하는 데이터프레임의 형식으로 바꿔주어야 함.

#새 데이터셋은=df.loc[행,x(df['일자]] 하동이거나 상동이면, 하구이거나 상구이면
#구랑 동 한 행마다 날짜 수만큼 * 리스트
#일자 일렬로 - *전체 행 수만큼
#그 안의 값 - 주의 ! 가로로 쭉쭉 세워서 붙여줘야 함.


goo=pp['구'].to_list()
dong=pp['동'].to_list()

gu=[]
dg=[]
dt=[]
num=[]

#구 * 일자
for i in goo:
  for j in range(len(pp.columns)-2):
    gu.append(i)

#동 * 일자
for i in dong:
  for j in range(len(pp.columns)-2):
    dg.append(i)

#일자 * 행 수
col=pp.columns[2:]
for i in range(pp.shape[0]):
  for j in col:
    dt.append(j)

#안의 값- 가로로 cum
for i in range(len(pp)):
  num.extend(pp.loc[i,'2017 1/4':].to_list())

gu=pd.DataFrame(gu)
dg=pd.DataFrame(dg)
dt=pd.DataFrame(dt)
num=pd.DataFrame(num)

peo=pd.concat([dt,gu,dg,num],axis=1)

peo.columns=['일자','구','동','인구']
peo
```

```python
#구랑 일자만으로 concat하려면 동 여러개 나오니까 소계만 있는 df 따로 빼야할듯
#남양주시,하남시,구리시 없음 ->#토평동(구리),삼패동(남양주),마포동,망월동(하남)

#여의도동->여의동 (명칭 변경으로 replace 처리 해주었음.)

peo['동'].replace({'여의동':'여의도동'},inplace=True)

peogu=peo[peo['동']=="소계"]

#"-"인 값 몇개 있는데 안 쓸 값이라 0으로 일단 대체
peo['인구'].replace({"-":"0"},inplace=True)

peogu
```

```python
#아예 결측치인 부분 처리(NaN값)
#상일1동은 있는 값의 평균, 상일 2동의 결측치는 10200 입력 후, 상승세 등차수열
#이촌동의 경우 경계가 모호해 이촌 1동, 2동의 평균으로 대체해주었음.

peo['인구']=peo['인구'].astype('int')

echon=peo.loc[peo['동'].str.contains("이촌")].groupby(['일자','구']).mean().astype('int').reset_index()
echon['동']="이촌동"
echon=echon[['일자','구','동','인구']] #순서 맞춰바꾸기 그냥 불러들이면 됨.
peo=pd.concat([peo,echon],axis=0) #이촌동을 생성해 밑에 붙여줌.
```

```python
#마포->용강도화, 신천동 -> 잠실 4,6
mapo=peo.loc[(peo['동'].str.contains("용강"))|(peo['동'].str.contains("도화"))].groupby(['일자','구']).mean().astype('int').reset_index()
mapo['동']="마포동"
mapo=mapo[['일자','구','동','인구']]
mapo

sc=peo.loc[(peo['동'].str.contains("잠실4"))|(peo['동'].str.contains("잠실6"))].groupby(['일자','구']).mean().astype('int').reset_index()
sc['동']="신천동"
sc=sc[['일자','구','동','인구']]
sc
```

```python
#기존의 df에 연결하기 위해 분기 기준을 일치시켜줌.
#1-> 그 연도
#2->1/4
#3->1/2
#4->3/4

df_p['기준_년분기_코드'].unique()
```

```python
def to_m(df_p):
  df_p['기준_년분기_코드']=df_p['기준_년분기_코드'].astype('str')

  df_p['분기']=df_p['기준_년분기_코드'].apply(lambda x: x[4])
  df_p['기준_년분기_코드']=df_p['기준_년분기_코드'].apply(lambda x: x[0:4])

  re={"1":"","2":"1/4","3":"1/2","4":"3/4"}
  df_p['분기'].replace(re,inplace=True)
  #df_p['분기']
  df_p['기준_년분기_코드']=df_p['기준_년분기_코드']+" "+df_p['분기']

  return df_p['기준_년분기_코드']
```

```python
df_p['기준_년분기_코드']=to_m(df_p)
df_p['기준_년분기_코드']
```

```python
#구 기준으로 먼저 넣기
df_pg['기준_년분기_코드']=to_m(df_pg)
df_pg['기준_년분기_코드']=df_pg['기준_년분기_코드'].str.rstrip()
df_pg['기준_년분기_코드']
```
